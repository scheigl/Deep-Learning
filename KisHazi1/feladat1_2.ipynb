```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Create an array for all five images with all three channels
images = np.zeros((5,128,128,3), dtype=np.uint8)

# Create an iterator for the loop
it = range(0,128)

# Create an iterator for the 5 images
repetition = range(0,5)

for k in repetition:

    # Create a 128x128 black images (arrays of pixels) for the three channels
    img_r = np.zeros((128,128,3), dtype=np.uint8)
    img_g = np.zeros((128,128,3), dtype=np.uint8)
    img_b = np.zeros((128,128,3), dtype=np.uint8)


    # Read the 'k'th image (and display it)
    img = cv2.imread('resources/image'+ str(k) +'.jpg')
    display("Picture "+ str(k) +": The original image")
    plt.imshow(img)
    plt.show()

    
    # The red components of the original image
    # We are copying only the red values from the original picture to the black image
    for i in it:
        for j in it:
            img_r[i,j,0] = img[i,j,0]
    # Plotting the red image
    display("Picture "+ str(k) +": The red image")
    plt.imshow(img_r)
    plt.show()
    
    # The red value matrix
    images[k,:,:,0] = img_r[:,:,0]
    display("Picture "+ str(k) +": The red matrix")
    display(img_r[:,:,0])

    
    # The green components of the original image
    # We are copying only the green values from the original picture to the black image
    for i in it:
        for j in it:
            img_g[i,j,1] = img[i,j,1]
    # Plotting the green image
    display("Picture "+ str(k) +": The green image")
    plt.imshow(img_g)
    plt.show()
    
    # The green value matrix
    images[k,:,:,1] = img_g[:,:,1]
    display("Picture "+ str(k) +": The green matrix")
    display(img_g[:,:,1])

    
    # The blue components of the original image
    # We are copying only the blue values from the original picture to the black image
    for i in it:
        for j in it:
            img_b[i,j,2] = img[i,j,2]
    # Plotting the blue image
    display("Picture "+ str(k) +": The blue image")
    plt.imshow(img_b)
    plt.show()

    # The blue value matrix
    images[k,:,:,2] = img_b[:,:,2]
    display("Picture "+ str(k) +": The blue matrix")
    display(img_b[:,:,2])
    
```

```python
# We create an other array for all the picutures
images_dist = np.zeros((5,128,128,3), dtype=float)

for k in repetition:
    
    # Average red value of the given image 
    average_r = np.average(images[k,:,:,0])
    display("Picture "+ str(k) +": The red matrix average: "+ str(average_r))

    # Standard deviation of the red value of the given image 
    deviation_r = np.std(images[k,:,:,0])
    display("Picture "+ str(k) +": The red matrix deviation: "+ str(deviation_r))

    # Standardising all the pixels of the red channel
    for i in it:
        for j in it:
            images_dist[k,i,j,0] = (images[k,i,j,0]-average_r)/deviation_r
    
    # Checking the answers
    display("Picture "+ str(k) +": The standardised red matrix average: "+ str(np.average(images_dist[k,:,:,0])))
    display("Picture "+ str(k) +": The standardised red matrix deviation: "+ str(np.std(images_dist[k,:,:,0])))
    
    print("\n")
    
    # Average green value of the given image 
    average_g = np.average(images[k,:,:,1])
    display("Picture "+ str(k) +": The green matrix average: "+ str(average_g))

    # Standard deviation of the green value of the given image 
    deviation_g = np.std(images[k,:,:,1])
    display("Picture "+ str(k) +": The green matrix deviation: "+ str(deviation_g))

    # Standardising all the pixels of the green channel
    for i in it:
        for j in it:
            images_dist[k,i,j,1] = (images[k,i,j,1]-average_g)/deviation_g
    
    # Checking the answers
    display("Picture "+ str(k) +": The standardised green matrix average: "+ str(np.average(images_dist[k,:,:,1])))
    display("Picture "+ str(k) +": The standardised green matrix deviation: "+ str(np.std(images_dist[k,:,:,1])))
    
    print("\n")
    
    # Average blue value of the given image 
    average_b = np.average(images[k,:,:,2])
    display("Picture "+ str(k) +": The blue matrix average: "+ str(average_b))

    # Standard deviation of the blue value of the given image 
    deviation_b = np.std(images[k,:,:,2])
    display("Picture "+ str(k) +": The blue matrix deviation: "+ str(deviation_b))

    # Standardising all the pixels of the blue channel
    for i in it:
        for j in it:
            images_dist[k,i,j,2] = (images[k,i,j,2]-average_b)/deviation_b
    
    # Checking the answers
    display("Picture "+ str(k) +": The standardised blue matrix average: "+ str(np.average(images_dist[k,:,:,2])))
    display("Picture "+ str(k) +": The standardised blue matrix deviation: "+ str(np.std(images_dist[k,:,:,2])))
    
    print("\n")
```
