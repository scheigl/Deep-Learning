```python
from bs4 import BeautifulSoup
import requests
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
```

```python
# HTML request from the webpage
result = requests.get("https://blog.keras.io/the-future-of-deep-learning.html")
```

```python
# Store our HTML content in a variable: soup
p = result.content
soup = BeautifulSoup(p)
```

```python
# We join all literal data from <p> blocks of the webpage
text = "".join([p.text for p in soup.find_all("p")])
```

```python
# We filter out all non-alphabetical characters, and keep only the letters
mixedletters = ''.join(filter(str.isalpha, text))
```

```python
# We change all the letters to lower case
letters = mixedletters.lower()
print(letters)
```

```python
# We count the letters and put it in a dataset
labels, values = zip(*Counter(letters).items())
# We create the histogram
indexes = np.arange(len(labels))
width = 1
plt.bar(indexes, values, width)
plt.xticks(indexes, labels)
plt.show()
```
